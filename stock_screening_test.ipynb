```python
import pandas as pd
import pandas_datareader as web
import matplotlib.pyplot as plt
import datetime
import yfinance as yf
```

```python
# 特定銘柄の情報を取得（Tは東証）
ticker_info = yf.Ticker("9984.T")

# 会社概要(info)を出力
ticker_info.info
```

```python
#春銘柄候補
ticker_list = [9983,8227,8016,2685,8011,2269,4151,4508,4506,4507,4523,3105,3107,3109,3514,3593,3604,
                4967,8113,2269,4452,4912,4527,4536,8889,8850,9039,9064,9831,8282,3048,3050,8218,8184,
                9843,8186,2201,2269,2217,2211,3382,8267,2695,2801,2695,9474]
#抽出する指標
info_label = ['ticker','shortName','website','revenueGrowth','earningsGrowth','returnOnAssets','returnOnEquity','revenuePerShare','priceToBook','trailingPE']
#yahoo financeからデータを抽出しdataframeで結合・出力
df_v = pd.DataFrame()
for ticker in ticker_list:
    ticker_info = yf.Ticker(str(ticker)+".T")
    d=ticker_info.info
    try:
        d["ticker"] = ticker
        df = pd.DataFrame(d.values(), index=d.keys()).T[info_label]
        df_v = pd.concat([df_v, df], axis=0)
    except:
        continue
df_v
```

```python
#行名列名変更
df_v = df_v.set_axis(['ticker','Name','website','revenueGrowth','earningsGrowth','ROA','ROE','EPS','PBR','PER'], axis=1)
df_v = df_v.set_axis(range(1,len(df_v)+1,1), axis=0)
df_v
```

```python
#ROA,ROEともに5％以上
df_v = df_v[df_v['ROA'] > 0.05]
df_v = df_v[df_v['ROE'] > 0.05]
#2022年に売上、利益が増加しているか
df_v = df_v[df_v['revenueGrowth'] > 0]
df_v = df_v[df_v['earningsGrowth'] > 0]
df_v
```

```python
#4年連続売上が増加している銘柄のみスクリーニング
for ticker in df_v["ticker"]:
    ticker_info = yf.Ticker(str(ticker)+".T")
    d=pd.DataFrame(ticker_info.earnings)
    d_Rsort = d.sort_values('Revenue')
    if d.equals(d_Rsort) == True:
        print(ticker)
```

```python
#4年連続利益が増加している銘柄のみスクリーニング
for ticker in df_v["ticker"]:
    ticker_info = yf.Ticker(str(ticker)+".T")
    d=pd.DataFrame(ticker_info.earnings)
    d_Esort = d.sort_values('Earnings')
    if d.equals(d_Esort) == True:
        print(ticker)
```
